# Carbon intensity data from National Grid

```python
import pandas as pd
```

```python
df = pd.read_csv('carbon_intensity_data.csv', index_col=0)[['CARBON_INTENSITY']]
df.index = pd.to_datetime(df.index)
df.rename(columns={'CARBON_INTENSITY': 'Carbon Emissions (gCO2/kWh)'}, inplace=True)
df
```

Check for missing time indices...

```python
pd.date_range(start='2009-01-01', end='2022-04-01').difference(df.index)
```

```python
df.plot(figsize=(15,10), title='Half-hourly Carbon Emissions (gCO2/kWh) over entire dataset')
```

```python
start='2021-01-01'
end='2021-12-31'

df[start:end].plot(figsize=(15,10), title='Half-hourly Carbon Emissions (gCO2/kWh) over 2021')
```

```python
start = '2021-09-01'
end = '2021-09-30'

df[start:end].plot(figsize=(15,10), title='Half-hourly Carbon Emissions (gCO2/kWh) over September 2021')
```

## Regional carbon intensity forecast

```python
df_regional = pd.read_csv('regional_carbon_intensity.csv', index_col=0)
df.index = pd.to_datetime(df.index)
df_regional
```

Select region

```python
region = 'South Scotland'

df_region = df_regional[[region]]
```

```python
title = region + ' Regional Half-hourly Carbon Emissions (gCO2/kWh) over entire dataset'
df_region.plot(figsize=(15,10), title=title)
```

```python
start='2021-01-01'
end='2021-12-31'

title = region + ' Regional Half-hourly Carbon Emissions (gCO2/kWh) over 2021'
df_region[start:end].plot(figsize=(15,10), title=title)
```

```python
start='2021-09-01'
end='2021-09-30'

title = region + ' Regional Half-hourly Carbon Emissions (gCO2/kWh) over September 2021'
df_region[start:end].plot(figsize=(15,10), title=title)
```
