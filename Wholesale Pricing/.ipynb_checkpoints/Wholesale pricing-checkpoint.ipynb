# Wholesale pricing calculation from Octopus tariffs

This notebook calculates wholesale prices using a back calculation from the Octopus agile and agile outgoing tariffs. It follows the method described in https://www.energy-stats.uk/wholesale-energy-pricing/.

```python
import pandas as pd
import numpy as np
```

Select from the following regions.

```python
regions = ['A_Eastern_England', 'B_East_Midlands', 'C_London', 'D_Merseyside_and_Northern_Wales',
           'E_West_Midlands', 'F_North_Eastern_England', 'G_North_Western_England', 'H_Southern_England',
           'J_South_Eastern_England', 'K_Southern_Wales', 'L_South_Western_England', 'M_Yorkshire',
           'N_Southern_Scotland', 'P_Northern_Scotland']

region = 'N_Southern_Scotland'
```

```python
filename_agile = 'csv_agile_' + region + '.csv'
df_import = pd.read_csv(filename_agile, index_col=0, header=None)#[['CARBON_INTENSITY']]
df_import.index = pd.to_datetime(df_import.index)
df_import.drop(columns=[1, 2, 3], inplace=True)
df_import.rename(columns={4: 'Agile price (p/kWh)'}, inplace=True)
df_import.index.name = 'Datetime'
df_import.plot(figsize=(15,10), title='Agile import price (p/kWh) over entire dataset')
```

Now we use the agile import price to back calculate the wholesale price. Required multipliers and adders (for 4-7pm) for equation caclulating Agile price is from https://www.guylipman.com/octopus/formulas. The agile import prices can only give us indicative wholesale prices for times where the cap of 35p/kWh is not reached.

To calculate the retail price from wholesale market data for your region (we are doing the following method backwards):

- Multiply the wholesale price in p/kWh by the multiplier
- For any periods from 4-7pm local time, add the adder
- Multiply the price by 1.05 to include VAT (we are not including this step)
- Cap the resultant price at 35p/kWh

```python
region_multiplier_dict = {'A_Eastern_England': 2.1, 'B_East_Midlands': 2.0,
                          'C_London': 2.0, 'D_Merseyside_and_Northern_Wales': 2.2,
                          'E_West_Midlands': 2.1, 'F_North_Eastern_England': 2.1,
                          'G_North_Western_England': 2.1, 'H_Southern_England': 2.1,
                          'J_South_Eastern_England': 2.2, 'K_Southern_Wales': 2.2,
                          'L_South_Western_England': 2.3, 'M_Yorkshire': 2.0,
                          'N_Southern_Scotland': 2.1, 'P_Northern_Scotland': 2.4}

region_adder_dict = {'A_Eastern_England': 13, 'B_East_Midlands': 14,
                     'C_London': 12, 'D_Merseyside_and_Northern_Wales': 13,
                     'E_West_Midlands': 12, 'F_North_Eastern_England': 12,
                     'G_North_Western_England': 12, 'H_Southern_England': 12,
                     'J_South_Eastern_England': 12, 'K_Southern_Wales': 12,
                     'L_South_Western_England': 11, 'M_Yorkshire': 13,
                     'N_Southern_Scotland': 13, 'P_Northern_Scotland': 12}

multiplier = region_multiplier_dict[region]
adder = region_adder_dict[region]

# adder is only between 4pm and 7pm
df_import['adder'] = df_import['Agile price (p/kWh)'].between_time('16:00', '19:00')
# change adder price to specified for region
mask = df_import["adder"].notna()
df_import.loc[mask, "adder"] = adder
# new column with adder price subtracted
df_import['Wholesale price (p/kWh) - Import'] = df_import['Agile price (p/kWh)'] - df_import['adder'].fillna(0) 
# drop the adder column
df_import.drop(columns=['adder'], inplace=True)

# divde by the multiplier to get wholesale price
df_import['Wholesale price (p/kWh) - Import'] /= multiplier

# start = '2021-09-01'
# end = '2021-09-01'
# df_import[start:end]
```

Now to look at the agile outgoing tariff in order to calculate wholesale prices when the Agile cap of 35p/kWh is reached.

To calculate the retail price for your region (again done backwards):

- Multiply the wholesale price in p/kWh by the multiplier shown
- For all periods, add the first adder shown
- For any periods from 4-7pm local time, add the second adder shown
- (There is no VAT on exported power)
- Floor the resultant price at 0p/kWh

```python
filename_agileoutgoing = 'csv_agileoutgoing_' + region + '.csv'
df_export = pd.read_csv(filename_agileoutgoing, index_col=0, header=None)
df_export.index = pd.to_datetime(df_export.index)
df_export.drop(columns=[1, 2, 3], inplace=True)
df_export.rename(columns={4: 'Agile export price (p/kWh)'}, inplace=True)
df_export.index.name = 'Datetime'
df_export.plot(figsize=(15,10), title='Agile export price (p/kWh) over entire dataset')
```

```python
region_multiplier_dict_export = {'A_Eastern_England': 0.95, 'B_East_Midlands': 0.94,
                                 'C_London': 0.95, 'D_Merseyside_and_Northern_Wales': 0.97,
                                 'E_West_Midlands': 0.94, 'F_North_Eastern_England': 0.95,
                                 'G_North_Western_England': 0.96, 'H_Southern_England': 0.94,
                                 'J_South_Eastern_England': 0.94, 'K_Southern_Wales': 0.94,
                                 'L_South_Western_England': 0.93, 'M_Yorkshire': 0.96,
                                 'N_Southern_Scotland': 0.97, 'P_Northern_Scotland': 0.96}

region_adder_dict_export = {'A_Eastern_England': 7.04, 'B_East_Midlands': 6.27,
                            'C_London': 5.93, 'D_Merseyside_and_Northern_Wales': 5.97,
                            'E_West_Midlands': 6.50, 'F_North_Eastern_England': 4.88,
                            'G_North_Western_England': 5.89, 'H_Southern_England': 7.05,
                            'J_South_Eastern_England': 7.41, 'K_Southern_Wales': 5.46,
                            'L_South_Western_England': 7.14, 'M_Yorkshire': 5.78,
                            'N_Southern_Scotland': 3.85, 'P_Northern_Scotland': 2.68}

region_adder_all_day_dict_export = {'A_Eastern_England': 1.09, 'B_East_Midlands': 0.78,
                                    'C_London': 1.30, 'D_Merseyside_and_Northern_Wales': 1.26,
                                    'E_West_Midlands': 0.77, 'F_North_Eastern_England': 0.87,
                                    'G_North_Western_England': 1.10, 'H_Southern_England': 0.93,
                                    'J_South_Eastern_England': 1.09, 'K_Southern_Wales': 0.97,
                                    'L_South_Western_England': 0.83, 'M_Yorkshire': 0.72,
                                    'N_Southern_Scotland': 0.90, 'P_Northern_Scotland': 1.36}

multiplier = region_multiplier_dict_export[region]
adder = region_adder_dict_export[region]
adder_all_day = region_adder_all_day_dict_export[region]

# adder is only between 4pm and 7pm
df_export['adder'] = df_export['Agile export price (p/kWh)'].between_time('16:00', '19:00')
# change adder price to specified for region
mask = df_export["adder"].notna()
df_export.loc[mask, "adder"] = adder
# new column with adder price subtracted
df_export['Wholesale price (p/kWh) - Export'] = df_export['Agile export price (p/kWh)'] - df_export['adder'].fillna(0) 
# drop the adder column
df_export.drop(columns=['adder'], inplace=True)

# subtract the all day adder
df_export['Wholesale price (p/kWh) - Export'] -= adder_all_day

# divde by the multiplier to get wholesale price
df_export['Wholesale price (p/kWh) - Export'] /= multiplier

# concat the dataframes
df = pd.concat([df_import, df_export], axis=1)

# start = '2021-09-01'
# end = '2021-09-01'
# df[start:end]
```

We are going to assume that Agile Import Tariff can be used to calculate wholesale price, except in periods where the cap is reached. For these periods we use the Agile Export tariff to calculate the wholesale price.

```python
df['Wholesale price (p/kWh)'] = np.where(df['Agile price (p/kWh)'] > 34.9, df['Wholesale price (p/kWh) - Export'], df['Wholesale price (p/kWh) - Import'])
start = '2021-09-01'
end = '2021-09-01'
df[start:end]
```
